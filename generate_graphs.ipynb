Generate accuracy and loss plots for the training and validation sets from csv files exported from tensorboard.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the csv files
train_acc = pd.read_csv('acc_train.csv')
test_acc = pd.read_csv('acc_test.csv')
train_loss = pd.read_csv('train_loss.csv')
test_loss = pd.read_csv('test_loss.csv')
```

```python
# Plot the accuracy
fc = 20
plt.rcParams['grid.alpha'] = 0.5

plt.figure(figsize=(16, 10))
plt.plot(train_acc['Step'], train_acc['Value'], label='Train', linewidth=2, color='blue')
plt.plot(test_acc['Step'], test_acc['Value'], label='Test', linewidth=2, color='red')
plt.title('Accuracy', fontsize=fc+5)
plt.xlabel('Epoch', fontsize=fc)
plt.ylabel('Accuracy', fontsize=fc)
plt.xticks(fontsize=fc)
plt.yticks(fontsize=fc)
plt.legend(fontsize=fc)
plt.grid()
plt.show()
```

```python
# Plot the loss
plt.figure(figsize=(16, 10))
plt.plot(train_loss['Step'], train_loss['Value'], label='Train', linewidth=2, color='blue')
plt.plot(test_loss['Step'], test_loss['Value'], label='Test', linewidth=2, color='red')
plt.title('Loss', fontsize=fc+5)
plt.xlabel('Epoch', fontsize=fc)
plt.ylabel('Loss', fontsize=fc)
plt.xticks(fontsize=fc)
plt.yticks(fontsize=fc)
plt.legend(fontsize=fc)
plt.grid()
plt.show()
```
